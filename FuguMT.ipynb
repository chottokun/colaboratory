```python
from transformers import pipeline
fugu_translator = pipeline('translation', model='staka/fugumt-en-ja')
fugu_translator('This is a cat.')
```

```python
text = "I consider that a man’s brain originally is like a little empty attic, and you have to stock it with such furniture as you choose."
```

```python
fugu_translator(text)
```
