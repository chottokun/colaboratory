```python
import pandas as pd
import pyLDAvis
import nlplot

def prepare_data():
    target_col = "text"
    texts = ["Think rich look poor",
             "When you come to a roadblock, take a detour",
             "When it is dark enough, you can see the stars",
             "Never let your memories be greater than your dreams",
             "Victory is sweetest when you’ve known defeat"]
    return pd.DataFrame({target_col: texts})

df = prepare_data()
```

```python
npt = nlplot.NLPlot(df, target_col='text')
```

```python
stopwords = npt.get_stopword(top_n=1, min_freq=0)
```

```python
npt.bar_ngram(
    title='uni-gram',
    xaxis_label='word_count',
    yaxis_label='word',
    ngram=1,
    top_n=50,
    stopwords=stopwords,
)
```

```python
# ビルド（データ件数によっては処理に時間を要します）
npt.build_graph(stopwords=stopwords, min_edge_frequency=0)

npt.co_network(
    title='Co-occurrence network',
)
```

```python
npt.wordcloud(
    max_words=100,
    max_font_size=100,
    colormap='tab20_r',
    stopwords=stopwords,
)
```

```python
npt.ldavis(num_topics=2, passes=5, save=True)
```
