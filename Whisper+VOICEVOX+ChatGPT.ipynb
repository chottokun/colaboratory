```python
# https://www.inoue-kobo.com/ai_ml/whisper/　を参考にさせていただきました。
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode

RECORD = """
const sleep = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=5, filename='temp/rec_audio.wav'):
  display(Javascript(RECORD))
  s = output.eval_js('record(%d)' % (sec * 1000))
  b = b64decode(s.split(',')[1])
  with open(filename, 'wb+') as f:
    print(filename)
    f.write(b)
```

## 動作確認

```python
import whisper

AUDIO_FILE="/content/temp/rec_audio.wav"

def getAudio(audiofile='temp/rec_audio.wav', second = 5):
  print(f'Speak to your microphone {second} sec...')
  record(sec = second, filename=audiofile)
  print('Done!')

# getAudio(audiofile=AUDIO_FILE)
```

```python
# https://zenn.dev/tam_tam/articles/d59250ecf25628
import whisper

model = whisper.load_model("base")

def getTextfromAudio(filename="temp/rec_audio.wav", lang="ja", model=model): 

  # Load audio
  audio = whisper.load_audio(filename)
  audio = whisper.pad_or_trim(audio)

  mel = whisper.log_mel_spectrogram(audio).to(model.device)

  # Output the recognized text
  options = whisper.DecodingOptions(language=lang, without_timestamps=True, fp16 = False)
  result = whisper.decode(model, mel, options)
  # print(result.text)

  return result.text

# print(getTextfromAudio(filename=AUDIO_FILE))
```

```python
#ChatGPT

def getMessage(text, prompt="", history=""):
  model="gpt-3.5-turbo"
  messages = [
      {
        "role": "system",
        "content": prompt
      },
      {
        "role":"user",
        "content": text
       }
  ]
  completion = openai.ChatCompletion.create(model=model,  messages=messages)
  text = str(completion.choices[0].message.content).replace("\n", "。")
  # print(text)
  return text

prompt="元気な高校生として日本語で答えてください。一人称は「僕」。"
text = "あなたの一番気になることを一行で答えてください。"
print(getMessage(text,prompt))
```

```python
#VOICEVOXの実行
from IPython.display import Audio
import subprocess

def talkText(text, speaker_id = 0):
  cmd_sed = ["sed", "-i", r"/SPEAKER_ID =/c SPEAKER_ID = "+str(speaker_id), r"/content/voicevox_core/run.py"]
  result = subprocess.run(cmd_sed, capture_output=True, text=True)
  cmd_talk = ["python","/content/voicevox_core/run.py","/content/voicevox_core/open_jtalk_dic_utf_8-1.11", text, "/content/data.wav"]
  result = subprocess.run(cmd_talk, capture_output=True, text=True)
  print(result.stdout)
  print(result.stderr) 
  # Audio("/content/data.wav", autoplay=True)
  return

%cd voicevox_core/
talkText(text = "起動しました。", speaker_id = 4)
Audio("/content/data.wav", autoplay=True)
```
