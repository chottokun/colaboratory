```python
import torch
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu') 
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "cerebras/Cerebras-GPT-1.3B"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

model.to(device)
```

```python
from transformers import pipeline

def generate_text(text, model=model, tokenizer=tokenizer, device=device):
  pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device=device)
  return pipe(text, max_length=200, do_sample=False, no_repeat_ngram_size=2)[0]

text = "Generative AI is "
print(generate_text(text, model)['generated_text'])
```

```python
%%time
text = "Thomas Edison is "
print(generate_text(text, model)['generated_text'])
```

```python
%%time
text = "人工知能とは、 "
print(generate_text(text)['generated_text'])
```
